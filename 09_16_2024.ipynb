#MELT

```python
df_melt = pd.melt(df, id_vars=['Rank','CCA3','Country/Territory','Capital','Continent','Area (km²)',
                               'Density (per km²)','Growth Rate','World Population Percentage'],

                  var_name='YEAR',
                  value_name='Total Population')
df_melt
```

#Replace , Strip

```python
df_melt ['Year'] = df_melt['YEAR'].str.replace('Population','', regex = False)
df_melt
```

```python
df_melt['Year'] = pd.to_datetime(df_melt['Year'])
df_melt['Total_Population'] = pd.to_numeric(df_melt['Total Population'])
df_melt

df_melt.drop(columns=['year'], inplace=True)
df_melt
```

```python
df_melt['Year'] = df_melt['Year'].apply(lambda x: x)
```
